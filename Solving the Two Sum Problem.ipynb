### _Can you please give me a the solution to a famous 2-sum problem? Also, can you give me three different approaches, the first should be the naive brute force approach, the second should be the approach using hashmaps, and the third one using the concept of sorting?_

I'll provide you with three different approaches to solve the famous Two Sum problem. The Two Sum problem asks you to find two numbers in an array that add up to a specific target sum and return their indices.


````python
### Approach 1: Brute Force (Naive Approach)
** Time
Complexity: ** O(nÂ²)
** Space
Complexity: ** O(1)
````

````python
def two_sum_brute_force(nums, target):
    """
    Brute force approach: Check all possible pairs
    """
    n = len(nums)
    for i in range(n):
        for j in range(i + 1, n):
            if nums[i] + nums[j] == target:
                return [i, j]
    return []  # No solution found


# Test the brute force approach
nums1 = [2, 7, 11, 15]
target1 = 9
result1 = two_sum_brute_force(nums1, target1)
print(f"Brute Force - Input: {nums1}, Target: {target1}")
print(f"Output: {result1} (indices of {nums1[result1[0]]} and {nums1[result1[1]]})")
````

````python
### Approach 2: Hash Map (Optimal Approach)
**Time Complexity:** O(n)
**Space Complexity:** O(n)
````

````python
def two_sum_hashmap(nums, target):
    """
    Hash map approach: Store complements in a dictionary
    """
    num_to_index = {}

    for i, num in enumerate(nums):
        complement = target - num
        if complement in num_to_index:
            return [num_to_index[complement], i]
        num_to_index[num] = i

    return []  # No solution found

# Test the hash map approach
nums2 = [3, 2, 4]
target2 = 6
result2 = two_sum_hashmap(nums2, target2)
print(f"Hash Map - Input: {nums2}, Target: {target2}")
print(f"Output: {result2} (indices of {nums2[result2[0]]} and {nums2[result2[1]]})")
````

````python
### Approach 3: Two Pointers with Sorting
**Time Complexity:** O(n log n)
**Space Complexity:** O(n)

**Note:** This approach modifies the original order, so we need to keep track of original indices.
````

````python
def two_sum_sorting(nums, target):
    """
    Sorting + Two pointers approach: Sort the array and use two pointers
    """
    # Create pairs of (value, original_index)
    indexed_nums = [(num, i) for i, num in enumerate(nums)]

    # Sort by value
    indexed_nums.sort(key=lambda x: x[0])

    left = 0
    right = len(indexed_nums) - 1

    while left < right:
        current_sum = indexed_nums[left][0] + indexed_nums[right][0]

        if current_sum == target:
            # Return original indices in sorted order
            idx1, idx2 = indexed_nums[left][1], indexed_nums[right][1]
            return sorted([idx1, idx2])
        elif current_sum < target:
            left += 1
        else:
            right -= 1

    return []  # No solution found

# Test the sorting approach
nums3 = [3, 2, 3]
target3 = 6
result3 = two_sum_sorting(nums3, target3)
print(f"Sorting - Input: {nums3}, Target: {target3}")
print(f"Output: {result3} (indices of {nums3[result3[0]]} and {nums3[result3[1]]})")
````

## Performance Comparison


````python
import time
import random

def performance_test():
    """Compare performance of all three approaches"""
    # Generate test data
    test_size = 1000
    nums = [random.randint(1, 100) for _ in range(test_size)]
    target = nums[0] + nums[1]  # Ensure there's a solution

    approaches = [
        ("Brute Force", two_sum_brute_force),
        ("Hash Map", two_sum_hashmap),
        ("Sorting", two_sum_sorting)
    ]

    print(f"Performance test with array size: {test_size}")
    print("-" * 50)

    for name, func in approaches:
        start_time = time.time()
        result = func(nums.copy(), target)
        end_time = time.time()

        print(f"{name:12}: {end_time - start_time:.6f} seconds - Result: {result}")

performance_test()
````

## Summary

1. **Brute Force**: Simple but inefficient for large datasets. Checks every possible pair.
2. **Hash Map**: Most efficient solution. Uses extra space to achieve linear time complexity.
3. **Sorting + Two Pointers**: Good balance between time and space, but modifies the original order.

**Recommendation**: Use the Hash Map approach for most cases as it provides the best time complexity O(n).

## Performance Testing with Different Array Sizes

Let
's see how each approach performs with different array sizes:

````python
import time
import random

def improved_performance_comparison():
    """
    Improved performance comparison with worst-case scenarios
    """
    print("IMPROVED PERFORMANCE COMPARISON")
    print("=" * 70)

    sizes = [100, 500, 1000, 2000, 5000]  # Added larger size
    approaches = [
        ("Brute Force", two_sum_brute_force),
        ("Hash Map", two_sum_hashmap),
        ("Sorting", two_sum_sorting)
    ]

    for size in sizes:
        print(f"\nArray size: {size}")
        print("-" * 40)

        # Create worst-case scenario for brute force
        nums = list(range(size))
        random.shuffle(nums)

        # Put the target pair at the END to make brute force work harder
        target = nums[-2] + nums[-1]  # Last two elements

        for name, func in approaches:
            # Run multiple times for better timing accuracy
            total_time = 0
            runs = 5

            for _ in range(runs):
                start_time = time.perf_counter()  # More precise timing
                result = func(nums.copy(), target)
                end_time = time.perf_counter()
                total_time += (end_time - start_time)

            avg_time = (total_time / runs) * 1000  # Convert to milliseconds
            print(f"{name:12}: {avg_time:8.3f}ms (avg of {runs} runs)")

improved_performance_comparison()
````
