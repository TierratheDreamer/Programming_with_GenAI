```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

We have imported:
- `PIL` (Python Imaging Library) for image processing
- `numpy` for numerical operations
- `matplotlib.pyplot` for displaying images




Let's create code to:
1. Load and display a sample image
2. Convert it to grayscale (black and white)
3. Display both original and converted images side by side

```python
# Load sample image - replace 'sample.jpg' with your image path
img = Image.open('test_image.png')
img_array = np.array(img)

# Convert to grayscale
img_gray = img.convert('L')
img_gray_array = np.array(img_gray)

# Display original and grayscale images
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

ax1.imshow(img_array)
ax1.set_title('Original Image')
ax1.axis('off')

ax2.imshow(img_gray_array, cmap='gray')
ax2.set_title('Grayscale Image')
ax2.axis('off')

plt.tight_layout()
plt.show()
```

The code above:
- Uses PIL to load and convert images
- Uses numpy to convert images to arrays for display
- Uses matplotlib to show images side by side
- The grayscale conversion is done using PIL's convert('L') method where 'L' mode means single channel grayscale

Note: Replace 'sample.jpg' with your actual image path


I'll create a function that converts a color image to grayscale using the PIL library's convert method, as demonstrated in the existing code.

```python
def convert_to_grayscale(image):
    """
    Convert a color image to grayscale
    Input: PIL Image object in RGB mode
    Output: PIL Image object in L (grayscale) mode
    """
    return image.convert('L')
```

I'll create a simple test image using NumPy and PIL that we can use for testing image processing code. We'll generate a synthetic image with some basic shapes and patterns.

```python
# Create a synthetic test image
img_size = (300, 400)
test_img = np.zeros(img_size + (3,), dtype=np.uint8)

# Add background
test_img[:, :] = [240, 248, 255]  # AliceBlue color

# Add a red rectangle
test_img[50:150, 100:300] = [255, 0, 0]

# Add a blue circle
y, x = np.ogrid[0:300, 0:400]
circle_mask = (x - 200) ** 2 + (y - 200) ** 2 <= 80 ** 2
test_img[circle_mask] = [0, 0, 255]

# Convert to PIL Image
test_image = Image.fromarray(test_img)

# Save the test image
test_image.save('test_image.png')

# Display the image
plt.figure(figsize=(8, 6))
plt.imshow(test_img)
plt.axis('off')
plt.title('Generated Test Image')
plt.show()
```

The code above creates a test image with:
- White background
- Red rectangle
- Blue circle
The image is saved as 'test_image.png' in the current directory and can be used for further image processing tests.
