### _Now on the sepal length feature I want to plot a kde plot using seaborn library.  How do I do it?_

To create a KDE (Kernel Density Estimation) plot for the sepal length feature using seaborn, you'll need to first import seaborn and then use the `sns.kdeplot()` function. Here's how you can do it:


```python
# Import seaborn for visualization
import seaborn as sns
import matplotlib.pyplot as plt

# Create a KDE plot for sepal length (SL)
plt.figure(figsize=(10, 6))
sns.kdeplot(data=iris_df, x='SL', fill=True)
plt.title('KDE Plot of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Density')
plt.grid(True, alpha=0.3)
plt.show()
```

If you want to create a more advanced KDE plot that shows the distribution for each species separately, you can use:

```python
# KDE plot with species differentiation
plt.figure(figsize=(12, 6))
ax = sns.kdeplot(data=iris_df, x='SL', hue='Species', fill=True, alpha=0.7)
plt.title('KDE Plot of Sepal Length by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Density')
plt.grid(True, alpha=0.3)
plt.show()
```

**Key parameters for `sns.kdeplot()`:**
- `data`: Your DataFrame (iris_df)
- `x`: The column you want to plot ('SL' for sepal length)
- `hue`: Optional - column to separate by categories ('Species')
- `fill`: Fills the area under the curve (True/False)
- `alpha`: Transparency level (0-1)

The KDE plot will show you the probability density distribution of sepal length values, helping you understand:
- The most common sepal length values (peaks)
- The spread of the data
- Whether the distribution is normal, skewed, or has multiple modes
- How different species compare (if using the `hue` parameter)

### _Can you provide also with the code to show all three species to separate plots, side by side, and also showing continuous curve?_

Absolutely! Here are several ways to create separate plots for each species side by side, including continuous curves:

## Method 1: Subplots with Histograms + KDE curves

```python
# Create subplots for each species with histogram and KDE curve
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
species_list = ['setosa', 'versicolor', 'virginica']
colors = ['blue', 'green', 'red']

for i, (species, color) in enumerate(zip(species_list, colors)):
    # Filter data for each species
    species_data = iris_df[iris_df['Species'] == species]['SL']

    # Create histogram
    axes[i].hist(species_data, bins=10, alpha=0.7, color=color, density=True, label='Histogram')

    # Add KDE curve
    sns.kdeplot(species_data, ax=axes[i], color='black', linewidth=2, label='KDE Curve')

    # Customize each subplot
    axes[i].set_title(f'{species.capitalize()} - Sepal Length Distribution')
    axes[i].set_xlabel('Sepal Length (cm)')
    axes[i].set_ylabel('Density')
    axes[i].legend()
    axes[i].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## Method 2: Using Seaborn FacetGrid (Recommended)

```python
# Create FacetGrid for separate species plots
g = sns.FacetGrid(iris_df, col='Species', col_wrap=3, height=4, aspect=1.2)

# Map histogram to each subplot
g.map(plt.hist, 'SL', bins=10, alpha=0.7, density=True)

# Add KDE curves
g.map(sns.kdeplot, 'SL', color='red', linewidth=2)

# Customize the plots
g.set_axis_labels('Sepal Length (cm)', 'Density')
g.set_titles('{col_name} Species')
for ax in g.axes.flat:
    ax.grid(True, alpha=0.3)

plt.show()
```

## Method 3: More Advanced with Statistics

```python
# Advanced version with statistical information
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
species_list = ['setosa', 'versicolor', 'virginica']
colors = ['skyblue', 'lightgreen', 'lightcoral']

for i, (species, color) in enumerate(zip(species_list, colors)):
    # Filter data for each species
    species_data = iris_df[iris_df['Species'] == species]['SL']

    # Create histogram
    axes[i].hist(species_data, bins=12, alpha=0.7, color=color, density=True,
                edgecolor='black', label='Histogram')

    # Add KDE curve
    sns.kdeplot(species_data, ax=axes[i], color='darkred', linewidth=3, label='KDE Curve')

    # Add mean line
    mean_val = species_data.mean()
    axes[i].axvline(mean_val, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_val:.2f}')

    # Add median line
    median_val = species_data.median()
    axes[i].axvline(median_val, color='orange', linestyle=':', linewidth=2, label=f'Median: {median_val:.2f}')

    # Customize each subplot
    axes[i].set_title(f'{species.capitalize()} Species\n(n={len(species_data)})', fontsize=14, fontweight='bold')
    axes[i].set_xlabel('Sepal Length (cm)', fontsize=12)
    axes[i].set_ylabel('Density', fontsize=12)
    axes[i].legend(fontsize=10)
    axes[i].grid(True, alpha=0.3)

    # Add statistical text
    stats_text = f'Mean: {mean_val:.2f}\nStd: {species_data.std():.2f}\nMin: {species_data.min():.2f}\nMax: {species_data.max():.2f}'
    axes[i].text(0.02, 0.98, stats_text, transform=axes[i].transAxes, fontsize=9,
                verticalalignment='top', bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

plt.tight_layout()
plt.show()
```

## Method 4: Seaborn with Multiple Plot Types


```python
# Create a more comprehensive visualization
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

species_list = ['setosa', 'versicolor', 'virginica']
colors = ['blue', 'green', 'red']

for i, (species, color) in enumerate(zip(species_list, colors)):
    species_data = iris_df[iris_df['Species'] == species]['SL']

    # Top row: Histogram with KDE
    sns.histplot(species_data, bins=12, alpha=0.7, color=color, kde=True,
                stat='density', ax=axes[0, i])
    axes[0, i].set_title(f'{species.capitalize()} - Histogram + KDE')
    axes[0, i].set_xlabel('Sepal Length (cm)')
    axes[0, i].set_ylabel('Density')
    axes[0, i].grid(True, alpha=0.3)

    # Bottom row: KDE only (smoother curve)
    sns.kdeplot(species_data, ax=axes[1, i], color=color, fill=True, alpha=0.7, linewidth=2)
    axes[1, i].set_title(f'{species.capitalize()} - KDE Only')
    axes[1, i].set_xlabel('Sepal Length (cm)')
    axes[1, i].set_ylabel('Density')
    axes[1, i].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

**Method 2** (Recommended):

**Method 4**:

## Enhanced Method 2: Improved Visuals with Minimal Code Changes


```python
# Enhanced FacetGrid with better visual appeal
g = sns.FacetGrid(iris_df, col='Species', col_wrap=3, height=4, aspect=1.2)

# Define colors for each species
colors = ['skyblue', 'lightgreen', 'lightcoral']
kde_colors = ['darkblue', 'darkgreen', 'darkred']

# Map histogram with species-specific colors and better separation
for i, (species, hist_color, kde_color) in enumerate(zip(['setosa', 'versicolor', 'virginica'], colors, kde_colors)):
    species_data = iris_df[iris_df['Species'] == species]['SL']

    # Get the corresponding axis
    ax = g.axes[i]

    # Create histogram with better visual separation
    ax.hist(species_data, bins=12, alpha=0.7, color=hist_color, density=True,
            edgecolor='black', linewidth=0.8)

    # Add KDE curve with distinct color
    sns.kdeplot(species_data, ax=ax, color=kde_color, linewidth=3)

# Customize the plots
g.set_axis_labels('Sepal Length (cm)', 'Density')
g.set_titles('{col_name} Species')
for ax in g.axes.flat:
    ax.grid(True, alpha=0.3)

plt.show()
```

```python
# Super simple enhanced version
g = sns.FacetGrid(iris_df, col='Species', col_wrap=3, height=4, aspect=1.2)

# Map histogram with edge color for better separation
g.map(plt.hist, 'SL', bins=12, alpha=0.7, density=True, edgecolor='black', linewidth=0.8)

# Add KDE curves with darker colors
g.map(sns.kdeplot, 'SL', color='black', linewidth=3)

# Customize
g.set_axis_labels('Sepal Length (cm)', 'Density')
g.set_titles('{col_name} Species')
for ax in g.axes.flat:
    ax.grid(True, alpha=0.3)

plt.show()
```

## Best Compromise: Simple + Species Colors


```python
# Best of both worlds: simple code + species-specific colors
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
species_list = ['setosa', 'versicolor', 'virginica']
colors = ['skyblue', 'lightgreen', 'lightcoral']
kde_colors = ['darkblue', 'darkgreen', 'darkred']

for i, (species, hist_color, kde_color) in enumerate(zip(species_list, colors, kde_colors)):
    species_data = iris_df[iris_df['Species'] == species]['SL']

    # Histogram with better edges
    axes[i].hist(species_data, bins=12, alpha=0.7, color=hist_color, density=True,
                edgecolor='black', linewidth=0.8)

    # KDE curve with distinct color
    sns.kdeplot(species_data, ax=axes[i], color=kde_color, linewidth=3)

    # Simple customization
    axes[i].set_title(f'{species.capitalize()} Species')
    axes[i].set_xlabel('Sepal Length (cm)')
    axes[i].set_ylabel('Density')
    axes[i].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

3. **Species-Specific Colors**: Each species gets its own color scheme for better readability
