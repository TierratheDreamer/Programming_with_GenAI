```python
# Import libraries

import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read the image 'house.jpeg'

def read_image(path_to_image):
    # Write your code to read the image from the file path
    image = cv2.imread(path_to_image)
    # Convert BGR to RGB for proper display with matplotlib
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    return image
```

```python
# Show the image
def show_image(image):
    # Write your code here
    plt.figure(figsize=(10, 8))
    plt.imshow(image, cmap='gray' if len(image.shape) == 2 else None)
    plt.axis('off')
    plt.title('Image')
    plt.show()
```

```python
# Generate edges of the image using Canny Edge Detector - a famous image edge detection technique!

def generate_canny_edge(image):
    # Write your code here
    # Convert to grayscale if the image is colored
    if len(image.shape) == 3:
        gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    else:
        gray = image

    # Apply Gaussian blur to reduce noise
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)

    # Apply Canny edge detection
    # Parameters: image, low_threshold, high_threshold
    edges = cv2.Canny(blurred, 50, 150)

    return edges
```

```python
# Load and process the image
image_path = 'house.jpeg'
original_image = read_image(image_path)

# Show original image
print("Original Image:")
show_image(original_image)

# Generate Canny edges
edges = generate_canny_edge(original_image)

# Show edge detected image
print("Canny Edge Detection Result:")
show_image(edges)

# Optional: Show both images side by side for comparison
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

axes[0].imshow(original_image)
axes[0].set_title('Original Image')
axes[0].axis('off')

axes[1].imshow(edges, cmap='gray')
axes[1].set_title('Canny Edge Detection')
axes[1].axis('off')

plt.tight_layout()
plt.show()
```

Here's what each function does:
1. **`read_image(path_to_image)`**:
    - Uses `cv2.imread()` to read the image from the file path
    - Converts from BGR to RGB color space for proper display with matplotlib

2. **`show_image(image)`**:
    - Uses matplotlib to display the image
    - Automatically handles grayscale images by using the 'gray' colormap
    - Removes axes for cleaner display

3. **`generate_canny_edge(image)`**:
    - Converts the image to grayscale if it's colored
    - Applies Gaussian blur to reduce noise (kernel size 5x5)
    - Applies Canny edge detection with thresholds of 50 (low) and 150 (high)
    - Returns the edge-detected binary image

The code also includes a demonstration section that:
- Loads the 'house.jpeg' image
- Displays the original image
- Generates and displays the Canny edge detection result
- Shows a side-by-side comparison

The Canny edge detection algorithm works in several steps:
1. **Noise reduction** using Gaussian blur
2. **Gradient calculation** to find edge strength and direction
3. **Non-maximum suppression** to thin edges
4. **Double thresholding** to identify strong and weak edges
5. **Edge tracking by hysteresis** to finalize edges

You can adjust the threshold values (50, 150) in the `cv2.Canny()` function to fine-tune the edge detection sensitivity for your specific image.

```python
# Generate edges of the image using Canny Edge Detector - a famous image edge detection technique!

def generate_canny_edge(image):
    # Write your code here
    # Convert to grayscale if the image is colored
    if len(image.shape) == 3:
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    else:
        gray = image

    # Apply Gaussian blur to reduce noise
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)

    # Apply Canny edge detection
    # Parameters: image, low_threshold, high_threshold
    edges = cv2.Canny(blurred, 50, 150)

    return edges
```
